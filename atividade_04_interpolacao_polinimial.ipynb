## Interpolação Polinomial - Método de Lagrange

Aluno: Ezequiel Henrique Melo do Nascimento

```python
import numpy as np
import matplotlib.pyplot as plt


def interpolacao_lagrange(xp, x, y, grau):
  yp = 0
  for k in range(0, grau+1):
    p = 1
    for j in range(0, grau+1):
      if k != j:
        p = p*(xp - x[j]) / (x[k] - x[j])

    yp = yp + p * y[k]

  return yp


# Dados iniciais

x = [35, 40, 45]
y = [0.99818, 0.99828, 0.99849]

grau = 2
xp = 32.5
yp = interpolacao_lagrange(xp, x, y, grau)

print ( "O calor específico da água a", xp," graus celsius é de:" ,yp)
```

Plotagem do gráfico do calor específico pela temperatura em celsius

```python
xp_aux = np.linspace(30, 45, 100)
yp_aux = []

for j in range(0, len(xp_aux)):
  yp_aux.append(interpolacao_lagrange(xp_aux[j], x, y, grau))

t = np.arange(-1,-1,-1)
yt=[]

for i in t :
  yt.append(interpolacao_lagrange(x,y,i))

plt.plot(t,yt,'b-')
plt.plot(x,y,"ro")
plt.plot(xp,yp,'g*')
plt.plot(xp_aux, yp_aux)
plt.grid()
plt.show()
```
